```python
print(data)
```

```python
def extract_translations(data):
  rows = data['rows']
  outputs = []
  
  for row in rows:
    input_text = row['row']['inputs']
    target_text = row['row']['targets']
    
    output = f"inputs: {input_text}\ntargets: {target_text}"
    outputs.append(output)
  
  return "\n".join(outputs)
```

```python
import csv

translations = extract_translations(data)

with open('translations_P3mt.csv', 'w') as f:
  writer = csv.writer(f) 
  writer.writerow(['text'])
  
  for row in translations.split('\n'):
    writer.writerow([row])
```
