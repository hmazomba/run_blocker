```python
def capitalize_full_name(full_name):
    return full_name.title()
```

```python
def format_submission(submission):
    submitted_at = submission['submittedAt']
    values = submission['values']
    
    contact_name = ''
    company = ''
    phone_number = ''
    email = ''
    customer_type = ''
    desired_course_package = ''
    desired_course = ''
    preferred_currency = ''
    linkedin_profile = ''
    city = ''
    country = ''
    
    for value in values:
        name = value['name']
        value = value['value']
        
        if name == 'full_name':
            contact_name = capitalize_full_name(value)
        elif name == 'company':
            company = value
        elif name == 'phone':
            phone_number = value
        elif name == 'email':
            email = value
        elif name == 'customer_type':
            customer_type = value
        elif name == 'desired_course_package':
            desired_course_package = value
        elif name == 'desired_course':
            desired_course = value
        elif name == 'preferred_currency':
            preferred_currency = value
        elif name == 'linkedin_profile':
            linkedin_profile = value
        elif name == 'city':
            city = value
        elif name == 'country':
            country = value
    
    formatted_submission = f"{submitted_at}\t{contact_name}\t{company}\t{phone_number}\t{email}\t{customer_type}\t{desired_course_package}\t{desired_course}\t{preferred_currency}\t{linkedin_profile}\t{city}\t{country}"
    return formatted_submission
```

```python
def convert_submissions(submissions):
    converted_submissions = []
    
    for submission in submissions:
        formatted_submission = format_submission(submission)
        converted_submissions.append(formatted_submission)
    
    return converted_submissions
```
