```python
def clean_data(data):
  rows = data['rows'] 
  output = []

  for row in rows:
    en = row['row']['translation']['en']
    xh = row['row']['translation']['xh']
    
    output.append(f"translation: {{'en': '{en}', 'xh': '{xh}'}}")
  
  return "\n".join(output)

cleaned_data = clean_data(data)

print(cleaned_data)
```
