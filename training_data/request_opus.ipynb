```python
import requests
import csv
import random
import json
import time
```

```python
# Parameters
dataset = 'opus_memat'  
config = 'xh-en'
split = 'train'
num_requests = 10
rows_per_request = 100
delay_between_requests = 5  # Delay in seconds
```

```python
url = "https://datasets-server.huggingface.co/rows?dataset=opus_memat&config=xh-en&split=train&offset=0&limit=100"
response = requests.get(url)
data=response.json()

# Print API response 
print(response.json())
```

```python
def extract_translations(rows):
  translations = []
  for row in rows:
    translation = row['row']['translation']
    en = translation.get('en')
    xh = translation.get('xh')

    if en and xh:
      translation_text = f"en: {en}\nxh: {xh}"
      translations.append(translation_text)

  return translations
```

```python
# Make requests
translations = []
for i in range(num_requests):

  # Generate random offset
  offset = random.randint(0, 150000) 
  
  url = f"https://datasets-server.huggingface.co/rows?dataset={dataset}&config={config}&split={split}&offset={offset}&limit={rows_per_request}"

  response = requests.get(url)
  
  try:
    response_json = response.json()
    rows = response_json.get('rows', [])
    print(rows)
    
    # Extract translations
    request_translations = extract_translations(rows)
    translations.extend(request_translations)
  
  except json.JSONDecodeError:
    print("Error decoding JSON response. Please check the API response.")
    continue

  # Introduce delay
  time.sleep(delay_between_requests)
  
```

```python
# Write to CSV
if translations:
  with open('opus_memat_translations_3.csv', 'w') as f:
    writer = csv.writer(f)
    writer.writerow(['text'])  # Change the column name to 'text'
    writer.writerows([[translation] for translation in translations])
  print("Translations successfully written to translations.csv!")
else:
  print("No translations found. The CSV file will be empty.")
```
